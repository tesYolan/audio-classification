````python
def conv(ni, nf, ks=3, stride=1, act=True, padding=None):
    res = Conv2d(ni, nf, stride=stride, kernel_size=ks)
    if act: res = Sequential(res, ReLU())
    return res
````

````python
arch = Sequential(
    conv(1,24,ks=5,act=False),
    MaxPool2d((4,2)),
    ReLU(),
    conv(24,48,ks=5,act=False),
    MaxPool2d((4,2),),
    ReLU(),
    conv(48,48,ks=5,act=True),
    Flatten(),
    Linear(2400,64),
    ReLU(),
    Linear(64,10)
)
````
